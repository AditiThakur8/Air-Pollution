```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
trainData = pd.read_csv('C:/Users/DELL/Downloads/Train.csv')
testData = pd.read_csv('C:/Users/DELL/Downloads/Test.csv')
trainData
```

```python
#defining x and y
y = trainData['target']
x = trainData.drop(['target'],axis = 1)
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
output = model.predict(x)
mse_train = mean_squared_error(y, output)
r2_train = r2_score(y, output)
print(f"Train MSE: {mse_train}, Train R^2: {r2_train}")

prediction_test = model.predict(testData)
```

```python
plt.scatter(y,output)
plt.grid()
plt.xlabel('Actual Air Quality Index')
plt.ylabel('Predicted Air Quality Index')
plt.title('Actual vs Predicted on Training Data')
plt.show()
```
